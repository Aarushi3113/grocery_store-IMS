```python
import json
import time
```

### Uploading Json file 

```python
fd = open("inventory.json",'r')
r = fd.read()
fd.close()

inventory = json.loads(r)

sale = {}
record = json.dumps(sale)
fs = open('sales.json','w')
fs.write(record)
fs.close()

category_list = ['Dal','Flour','Oil','Chips','Biscuits','Soft_Drinks']
```

## User Driven Program

```python
ch = input("Are you the Owner(O)/Customer(C): ")

# function available to the owner
if(ch == 'O'):
    print("-----MAIN MENU-----")
    print("1. Display Inventory")
    print("2. Add to Inventory")
    print("3. Restock and update Inventory")
    print("4. Delete from Inventory")
    print("5. Display Sale records")
    print("\n\n")
    c = 'y'
    while(c=='y'):
        n = int(input("Enter your choice(1,2,3,4 or 5)"))
        
        if(n==1):
            #displaying Inventory
            for i in inventory:
                print(i,inventory[i])
        
        elif(n==2):
            #adding a new item to inventory
            fd = open("inventory.json", 'w')
            name = input("Enter the name of product: ")
            brand = input("Enter the brand of product: ")
            category = input("Enter the category of the product: ")
            quality = input("Enter the quality of product: ")
            price = int(input("Enter the cost of product: "))
            quantity = int(input("Enter the quantity of product: "))
            inventory[1000+ len(inventory)+1] = {'name':name,'brand':brand,'category':category,'quality':quality,'price':price,'quant':quantity}
            json.dump(inventory,fd)
            fd.close()
            print("Item added Successfully")
        
        elif(n==3):
            #updating the inventory
            fd = open("inventory.json","w")
            print("1. Restock")
            print("2. Update Price")
            ch = int(input("Enter your choice(1/2)"))
            if(ch==1):
                item = input("Enter the item id you want to restock: ")
                quant = int(input("Enter the quantity of product added: "))
                inventory[item]['quant'] += quant
                print("Stock added successfully")
            elif(ch==2):
                item = input("Enter the id of product whose price you want to change: ")
                price = int(input("Enter the new price: "))
                inventory[item]['price'] = price
                print("Updated product price")
            else:
                print("Invalid Choice")
            json.dump(inventory,fd)
            fd.close()
        
        elif(n==4):
            #deleting from inventory
            fd = open("inventory.json","w")
            item = int(input("Enter the product id you want to remove: "))
            rem = inventory.pop(item,'No product found')
            print('The id of key removed is {}'.format(str(rem)))
            json.dump(inventory,fd)
            fd.close()
        
        elif(n==5):
            #checking sales file
            fs = open("sales.json","r")
            sale = fs.read()
            fs.close()
            record = json.loads(sale)
            print("Sales Record")
            print(record)
        
        else:
            print("!!!Invalid Choice!!!")
        c = input("Do you want to continue(y/n)")    
    print("------------Thank You-----------")

#function available to the customer    
elif(ch=='C'):
    print("\n\t$$-----Welcome to our Store------$$")
    print("\t\tOur Invenory\n")
    for i in category_list:
        print("\tList of Food Items in {} category: ".format(i))
        print("----------------------------------------------------")
        print("P_id ---- Name ------ Brand ----- Size ----- Price\n")
        for j in inventory:
            if(i == inventory[j]['category']):
                print("{}  {}    {}   {}     Rs {}".format(j,inventory[j]['name'],inventory[j]['brand'],inventory[j]['quality'],inventory[j]['price']))
        print()
    n = int(input("How many items do you want to buy")) 
    f = open("receipt.txt","a")
    f.write('\n$$$Your order is$$$\n')
    f.write('Name --- Brand --- Category --- Size --- Cost_per_unit --- Quanity --- Cost\n')
    fd = open("inventory.json", 'w')
    total_bill = 0
    Time = time.ctime()
    for i in range(n):
        bill = 0
        pid = input("Enter the product id: ")
        quant = int(input("Enter the quantity: "))
        if pid in inventory.keys():
            print("Product entered: ", inventory[pid]['name'])
        else:
            print("Product not found")
        #checking if product is available
        if(quant<=inventory[pid]['quant']):
            bill = quant * inventory[pid]['price']
            inventory[pid]['quant'] = inventory[pid]['quant'] - quant
        else:
            print("Unsufficient stocks. Only {} remaining".format(inventory[pid]['quant']))
            quant = int(input("Enter new quanity"))
            bill = quant * inventory[pid]['price']
            inventory[pid]['quant'] = inventory[pid]['quant'] - quant
        total_bill += bill
        order = inventory[pid]['name']+" "+inventory[pid]['brand']+" "+inventory[pid]['category']+" "+inventory[pid]['quality']+" "+str(inventory[pid]['price'])+" "+str(quant)+" "+str(bill)+"\n"
        #updating receipt text file
        f.write(order)
        #updating the inventory
        json.dump(inventory,fd)
        #updating sales.json file
        sale[len(sale)+1] = {'name':inventory[pid]['name'], 'brand':inventory[pid]['brand'],'size':inventory[pid]['quality'],'cost_per_unit':inventory[pid]['price'],'quant':quant,'amt':bill,'time':Time}
    f.write("----------------------------------\n")    
    order = "Total bill is: Rs"+ str(total_bill) 
    f.write(order) 
    f.write('\n---------------------------------\n')
    f.close()
    fd.close()
    fs = open("sales.json","w")
    json.dump(sale,fs)
    fs.close()
    f = open("receipt.txt", 'r+')
    while(True):
        line = f.readline()
        if not line:
            break
        print("{}".format(line.strip()))    
    f.seek(0)
    f.truncate(0)
    f.close()
    print("\n\n")
    print("***********************************************************")
    print("\tTHANKS FOR VISITING!!!HAVE A GOOD DAY!!")
```
